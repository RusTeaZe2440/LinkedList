```python
class node:
    def __init__(self,val) -> None:
        self.val = val 
        self.next = None
```

```python
class linked:
    def __init__(self) -> None:
        self.head = None
        self.n = 0

    def __len__(self):
        return self.n
    
    def insert_head(self,value):
        newnode = node(value)
        newnode.next = self.head
        self.head = newnode
        self.n += 1

    def insert_tail(self,value):
        newnode = node(value)
        curr = self.head
        if self.head == None:
            self.head = newnode
            self.n += 1
            return
        while curr.next != None:
            curr = curr.next
        curr.next = newnode
        self.n += 1

    def insert_value(self,after,value):
        newnode = node(value)
        curr = self.head
        while curr:
            if curr.val == after:
                break
            curr = curr.next       
        if curr:
            newnode.next = curr.next
            curr.next = newnode
            self.n += 1

    def clear(self):
        self.head = None
        self.n = 0

    def delete_head(self):
        if self.head == None:
            return 'Empty Linked list'
        self.head = self.head.next
        self.n -= 1

    def delete_tail(self):
        curr = self.head 

        if self.head == None:
            return 'Empty Linked List'
        while curr.next.next:
            curr = curr.next
        
        curr.next = None
        self.n -= 1
    
    def remove(self,value):
        curr = self.head
        while curr.next:
            if curr.next.val == value:
                break
            curr = curr.next
        if curr.next == None:
            return 'Item not Found'
        else:
            curr.next = curr.next.next
        
        self.n -= 1
         
    def search(self,value):
        curr = self.head
        pos = 0
        while curr:
            if curr.val == value:
                return pos
            curr = curr.next
            pos += 1

    def search_index (self,index):
        curr = self.head
        pos = 0
        while curr:
            if pos == index:
                return curr.val
            curr = curr.next
            pos += 1

    def replace_max(self,value):
        curr = self.head
        max = self.head
        while curr:
            if curr.val > max.val:
                max = curr
            curr = curr.next
            
        max.val = value
    
    def replace_min(self,value):
        curr = self.head
        min = self.head
        while curr:
            if curr.val < min.val:
                min = curr
            curr = curr.next
        
        min.val = value
    
    def sum_odd_position(self):
        curr = self.head
        pos = 0
        sum = 0
        while curr:
            if pos % 2 == 1:
                sum += curr.val
            curr = curr.next
            pos += 1
        return sum
    
    def reverse_LinkedList(self):
        curr = self.head
        prev_node = None
        while curr:
            next_node = curr.next
            curr.next = prev_node
            prev_node = curr 
            curr = next_node
        
        self.head = prev_node
    
    def traverse(self):
        curr = self.head
        while curr:
            print(curr.val ,end=' ')
            curr = curr.next

    def change_sent(self):
        curr = self.head
        while curr:
            if curr.val == '*' or curr.val == '/':
                curr.val == ' '
                
                if curr.next.val == '*' or curr.next.val == '/':
                    curr.next.next.val = curr.next.next.val.upper()
                    curr.next = curr.next
        
            curr = curr.next

    def remove_duplicate(self):
        curr = self.head
        while curr and curr.next:
            if curr.val == curr.next.val:
                curr.next = curr.next.next
                self.remove_duplicate()
            curr = curr.next
       
    def __repr__(self) -> str:
        curr = self.head
        elements = []
        while curr:
            elements.append(str(curr.val))
            curr = curr.next
        return str('->'.join(elements))
```
